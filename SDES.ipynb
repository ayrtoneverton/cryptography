```python
FIXED_IP = [2, 6, 3, 1, 4, 8, 5, 7]
FIXED_EP = [4, 1, 2, 3, 2, 3, 4, 1]
FIXED_IP_INVERSE = [4, 1, 3, 5, 7, 2, 8, 6]
FIXED_P10 = [3, 5, 2, 7, 4, 10, 1, 9, 8, 6]
FIXED_P8 = [6, 3, 7, 4, 8, 5, 10, 9]
FIXED_P4 = [2, 4, 3, 1]

S0 = [[1, 0, 3, 2],
      [3, 2, 1, 0],
      [0, 2, 1, 3],
      [3, 1, 3, 2]]

S1 = [[0, 1, 2, 3],
      [2, 0, 1, 3],
      [3, 0, 1, 0],
      [2, 1, 0, 3]]

def permutate(original, fixed_key):
    new = ''
    for i in fixed_key:
        new += original[i - 1]
    return new

def left_half(bits):
    return bits[:len(bits)//2]

def right_half(bits):
    return bits[len(bits)//2:]

def shift(bits):
    rotated_left_half = left_half(bits)[1:] + left_half(bits)[0]
    rotated_right_half = right_half(bits)[1:] + right_half(bits)[0]
    return rotated_left_half + rotated_right_half

def key1():
    return permutate(shift(permutate(KEY, FIXED_P10)), FIXED_P8)

def key2():
    return permutate(shift(shift(shift(permutate(KEY, FIXED_P10)))), FIXED_P8)

def xor(bits, key):
    new = ''
    for bit, key_bit in zip(bits, key):
        new += str(((int(bit) + int(key_bit)) % 2))
    return new

def lookup_in_sbox(bits, sbox):
    row = int(bits[0] + bits[3], 2)
    col = int(bits[1] + bits[2], 2)
    return format(sbox[row][col], '02b')

def f_k(bits, key):
    L = left_half(bits)
    R = right_half(bits)
    bits = permutate(R, FIXED_EP)
    bits = xor(bits, key)
    bits = lookup_in_sbox(left_half(bits), S0) + lookup_in_sbox(right_half(bits), S1)
    bits = permutate(bits, FIXED_P4)
    return xor(bits, L)

def encrypt(plain_text):
    bits = permutate(plain_text, FIXED_IP)
    temp = f_k(bits, key1())
    bits = right_half(bits) + temp
    bits = f_k(bits, key2())
    res = permutate(bits + temp, FIXED_IP_INVERSE)
    return res

def decrypt(cipher_text):
    bits = permutate(cipher_text, FIXED_IP)
    temp = f_k(bits, key2())
    bits = right_half(bits) + temp
    bits = f_k(bits, key1())
    res = permutate(bits + temp, FIXED_IP_INVERSE)
    return res

def encrypt_string(text):
    view = ''
    for i in text:
        view += (chr(int(encrypt(format(ord(i), '08b')),2)))
    return view

def decrypt_string(aux):
    lk = []
    for i in aux:
        lk.append(chr(int(decrypt(i),2)))
    return ''.join(lk)

def string_to_list(text):
    return [format(ord(c), '08b') for c in text]

def SDES(key, text):
    global KEY
    KEY = format(int(key), '010b')
    return ''.join(encrypt_string(text))

def SDES_Decode(key, text):
    global KEY
    KEY = format(int(key), '010b')
    return decrypt_string(string_to_list(text))
```

```python
def print_result_hexadecimal(key, text):
    encrypted = SDES(key, text)
    decrypted = SDES_Decode(key, encrypted)
    encrypted = ''.join(['%02X' % ord(c) for c in encrypted])
    print()
    print('Plaintext: ', text)
    print('encrypted: ', encrypted)
    print('decrypted: ', decrypted)

print_result_hexadecimal('642','Technology')
print_result_hexadecimal('42','Technology')
print_result_hexadecimal('42','technology')
print_result_hexadecimal('395','Power Rangers')
print_result_hexadecimal('98','The big house')
```
